```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from config import CONFIG
```

```python
img_to_align_path = CONFIG.data / "imgs" / "f1040--1988-1.png"
img_to_refer_path = CONFIG.data / "imgs" / "test_img.png"
```

```python
img_to_align = cv2.imread(str(img_to_align_path))
img_to_refer = cv2.imread(str(img_to_refer_path))
```

```python
rh, rw, rc = img_to_refer.shape
ah, aw, ac = img_to_align.shape
```

```python
scaled_h = ah / rh
scaled_w = aw / rw
rescaled_ah = int(ah / scaled_h)
rescaled_aw = int(aw / scaled_w)
rescaled_ah, rescaled_aw, img_to_refer.shape
```

```python
dim = (rescaled_aw, rescaled_ah)
img_to_align = cv2.resize(img_to_align, dim, interpolation=cv2.INTER_AREA)
img_to_align.shape, img_to_refer.shape
```

```python
plt.figure(figsize=(15, 10))
ax = plt.subplot(1, 2, 1)
plt.imshow(img_to_align)
ax = plt.subplot(1, 2, 2)
plt.imshow(img_to_refer)
```
